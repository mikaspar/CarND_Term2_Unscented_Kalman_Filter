```python
import os
import glob
import numpy as np
from matplotlib import pyplot as plt
```

```python
nis_output = 'build/output_file.txt'
```

```python
lines = []

with open(nis_output) as file:
    for line in file: 
        line = line.strip()
        lines.append(line)
```

```python
time_stamp_R = []
time_stamp_L = []
nis_R        = []
nis_L        = []
i            = 0
j            = 0

for x in lines:
    y = x.split(",")
    if y[1] == 'R' : time_stamp_R.append(int(y[0])) ; nis_R.append(float(y[2]))
    if y[1] == 'L' : time_stamp_L.append(int(y[0])) ; nis_L.append(float(y[2]))   
        
y_points     = np.ones_like(nis_R)
border_3     = [i * 7.8 for i in y_points]  
border_2     = [j * 5.99 for j in y_points]        
```

```python
plt.plot(time_stamp_R,nis_R, 'b')
plt.plot(time_stamp_R,border_3, 'r')
plt.ylim((0,15))
plt.title('NIS Radar')
plt.show
plt.savefig('nis_radar.jpg')
```

```python
plt.plot(time_stamp_L,nis_L, 'b')
plt.plot(time_stamp_L,border_2, 'r')
plt.ylim((0,15))
plt.title('NIS Laser')
plt.show
plt.savefig('nis_laser.jpg')
```
